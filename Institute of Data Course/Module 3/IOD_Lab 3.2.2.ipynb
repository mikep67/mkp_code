##### 8. Calculate Standard Deviation


Because we are dealing with two samples we need to compute the joint standard deviation. 

Here sp is the pooled standard deviation for n = n1 = n2 and s^2
X1 and s^2 X2 are the unbiased estimators of the variances of the two samples.

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/f63e99c30c01445668ea2cce9832da0f6810cb4b)

```python
# ANSWER
# Refer: Pyython Standard Deviation Tutorial
# https://likegeeks.com/python-standard-deviation/#Pooled-standard-deviation

import statistics
# Sample1 = Large House Median Value
Sample1 = lhs["MEDV"]

# Sample2 = Small House Median Value
Sample2 = shs["MEDV"]

# Standard Deviation - Sample 1
sd1 = statistics.stdev(Sample1)

# Standard Deviation - Sample 2
sd2 = statistics.stdev(Sample2)

# Calculate the Pooled Standard Deviation
pooled_sd = math.sqrt(((len(Sample1)-1) * (sd1 ** 2) + (len(Sample2)-1) * (sd2 ** 2))/(len(Sample1) -1  ) + (len(Sample2) -1))

print(pooled_sd)
```

```python
# Pooled Standard Deviation - Alternative
import statistics
# Sample1 = Large House Median Value
Sample1 = lhs["MEDV"]

# Sample2 = Small House Median Value
Sample2 = shs["MEDV"]

# Standard Deviation - Sample 1
sd1 = statistics.stdev(Sample1)

# Standard Deviation - Sample 2
sd2 = statistics.stdev(Sample2)

# Calculate the Pooled Standard Deviation with the suggest formula

pooled_sd = math.sqrt(((sd1**2) + (sd2**2))/2 )
print(pooled_sd)
```

##### 9. Calculate test statistic

The *t*-statistic depends on the difference between the sample means and their joint standard deviation:

![](https://wikimedia.org/api/rest_v1/media/math/render/svg/070d365e1b347ea5f83f0147043868fa120b6646)


```python
# ANSWER
# Sample population size 
N  = 200
# Calculate the t-statistics
t = (Sample1.mean() - Sample2.mean())/(pooled_sd*np.sqrt(2/N))

print(t)
```

##### 10. Calculate Degree of Freedom

The number of degrees of freedom is the total number of samples (here, this is N from each set) minus the number of statistics in the computation (1 for each sample mean):

```python
# ANSWER
#Degrees of freedom differs from Excel T tests
df = 2*N - 2
print(df)
```

##### 11. Find the p-value

Now we can compute the probability of the above *t*-statistic from the cumulative density function:

> stats.t.cdf

> p_value = 1 - cdf(t) 

```python
# ANSWER#
# p-value after comparison with the t 

p_value = 1 - stats.t.cdf(t,df=df)
print(p_value)



```

#### Interpret results

##### 12. Write a Function to Test Significance of `p_value`

Write a funtion which will take p_value and alpha as input. If p_value < alpha, print reject null hypothesis message. Otherwise print a fail to reject null hypothesis message. 

```python
# ANSWER
# interpret

def hyp_test(alpha,p_value):

    if p_value < alpha:
	    return 'Reject null hypothesis'
    else:
	    return 'Failed to reject null hypothesis' 
```

```python
# ANSWER
alpha = int(0.05)

hyp_result = hyp_test(alpha, p_value)


print(hyp_result)
```
