## Categorical variables

#### Example 1: `MSZoning`

```python
# Count values in MSZoning

ames['MSZoning'].value_counts(dropna=False)
```

```python
# View first 10 rows of MSZoning

ames['MSZoning'][:10]
```

```python
# Strategy 1: dummy columns

mszones = pd.get_dummies(ames['MSZoning'], prefix='MSZone')
mszones.head(10)
```

```python
# Join with ames df

ames_2 = ames.join(mszones)
columns = ['MSZoning'] + mszones.columns.tolist()

ames_2[columns].head(10)
```

```python
# Drop least frequent zone to avoid redundancy

ames_2 = ames_2.drop('MSZone_C (all)', axis=1)
ames_2.columns
```

#### Example 2: `PoolQC`

```python
# Count values for PoolQC

ames['PoolQC'].value_counts(dropna=False)
```

```python
# View first 10 rows of PoolQC

ames['PoolQC'].head(10)
```

```python
def bin_pool(x):
    """If PoolQC is NaN, value is 0, else 1"""
    if pd.isnull(x):
        return 0
    else:
        return 1
```

```python
# Apply bin_pool to PoolQC and store result in new column Pool_Y
ames_2['Pool_Y'] = ames['PoolQC'].apply(bin_pool)

ames_2[['PoolQC', 'Pool_Y']].head(10)
```

```python
# Check values for Pool_Y

ames_2['Pool_Y'].value_counts(dropna=False)
```

```python
def rank_pool(x):
    """If PoolQC is NaN, value is 0, else rank quality"""
    if pd.isnull(x):
        return 0
    elif x == 'Fa':
        return 1
    elif x == 'Gd':
        return 3
    else:
        return 4
```

```python
# Rank PoolQC

ames_2['PoolQC_Num'] = ames_2['PoolQC'].apply(rank_pool)

ames_2['PoolQC_Num'].value_counts(dropna=False)
```
