### Import libraries

```python
# Import os functions
import os

# Import plotting libraries
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

#Import Numpy
import numpy as np

#Import Sci Learn Libraries
import sklearn
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split

# Keras Libraries with Tensorflow
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

### Load data
Load the Diabetes dataset from **SciKit-Learn**.

```python
# Load the diabetes 
diabetes = sklearn.datasets.load_diabetes(as_frame=True)
```

```python
diabetes.DESCR
```

```python
diabetes.data
```

```python
diabetes.data.columns
```

### Prepare input data

```python
X = diabetes.data.to_numpy()
n_cols = X.shape[1]
```

```python
y = diabetes.target.to_numpy()
```

### Split the data (training/test)

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size = 0.3,
    random_state = 65)
```

```python
# Set up the model architecture
model = Sequential()
```

### Create the model's architecture

```python
# Add the first hidden layer
model.add(Dense(100, activation = 'relu', input_shape = (n_cols, )))
```

```python
# Add the second hidden layer
model.add(Dense(50, activation = 'relu'))
```

```python
# Add the output layer
model.add(Dense(302, activation = 'softmax'))
```

### Compile the model

```python
# Compile the model
model.compile(
    optimizer = 'sgd',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy'])
```

```python
%%time
# Fit the model
history = model.fit(
    X_train,
    y_train,
    validation_split = 0.25,
    batch_size = 20,
    epochs = 100,
    verbose = 0)
```
